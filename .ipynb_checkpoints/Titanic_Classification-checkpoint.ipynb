# Titanic Classification

In this notebook, we are going to develop model to predict if a person can **survive or not** to the Titanic accident.

### > Data Collection

```python
# import librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_excel("datasets/titanic.xls")
orig_df = df
```

### > Data Cleaning

```python
df.dtypes
```

```python
# DataFrame shape
df.shape
```

The dataset contains 1309 records and 14 columns(caracteristics)

```python
# check missing values inside the dataset
df.isnull().mean()*100
```

```python
# Summarize the dataset directily using ydata-profiling librairie
from ydata_profiling import ProfileReport

ProfileReport(df)
```

According to the profilReport, we can draw some insightful observations:
 > * cabin, boat, body, home.dest columns have several missing values. So they can't help our model in the training process.
 > * name and ticket columns are not insightful. A model can't learn from this columns cause:
    1. all values from this columns are distincts. which doesn't help the model.
    2. this columns are leaky

```python
df = df.drop(columns = [
    "name",
    "ticket",
    "home.dest",
    "boat",
    "body",
    "cabin",
])
```

```python
df.head()
```

Now, we'll create new variables according to those that contains text values (sex, embarked)

```python
# we'll use the get_dummies parameter of pandas
df = pd.get_dummies(df, drop_first = True) # the drop first parameter means to drop first new variable created
df.head()
```

* For sex_male column, 0 means the person is female and 1 the person is male.
* For the embarked columns, it's the same process but when both are 0, it means embarked = **C** 

### > Data Splitting & Normalisation

```python
# X matrix contains caracteristics and y matrix contains target
X = df.drop(columns = "survived")
y = df["survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Well, now we've the train and the test set, we can fill the missing values.

```python
# Data Imputing
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

num_cols = ["pclass", "age", "sibsp", "parch", "fare"]

imputer = IterativeImputer()
# train set
imputed = imputer.fit_transform(X_train[num_cols])
X_train.loc[:, num_cols] = imputed
# test set
imputed = imputer.transform(X_test[num_cols])
X_test.loc[:, num_cols] = imputed
```

```python
# Data Normalisation
from sklearn.preprocessing import StandardScaler
cols = ['pclass', 'age', 'sibsp', 'parch', 'fare', 'sex_male',
       'embarked_Q', 'embarked_S']
sca = StandardScaler()
# train set
X_train = sca.fit_transform(X_train)
X_train = pd.DataFrame(X_train, columns = cols)
# test set
X_test = sca.transform(X_test)
X_test = pd.DataFrame(X_test, columns= cols)
```

### > Model Training & Evaluation

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Prediction
y_pred = rf.predict(X_test)
```

```python
# Model Evaluation
from sklearn.metrics import precision_score

p_score = precision_score(y_test, y_pred)
print(f"Prediction score > {p_score}")
```

For our model training, we use a tree based model. We can inspect the weights of the variables involve in the training process.

```python
for col, val in sorted(zip(X_train.columns, rf.feature_importances_,),
                       key= lambda x: x[1], reverse = True)[:5]:
    print(f"{col:10}{val:10.3f}")
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
from yellowbrick.classifier import ConfusionMatrix
%matplotlib inline

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('RandomForestClassifier Confusion Matrix')
plt.show()
```
